```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("pizzas.csv")
```

```python
df
```

```python
df.plot(kind='scatter', x='diametro', y='preco');
```

```python
from sklearn import linear_model
modelo = linear_model.LinearRegression()
x = df[["diametro"]]
y = df[["preco"]]
```

```python
modelo.fit(x, y)
```

```python
pred = modelo.predict([[30]])[0][0]
print(pred)
```
